```python
"""
Date: 10/21/2024 # Update the date after each code update
Course: COSC112.003_Fall 2024

Mid-Term project:
 Combat Arena Game- a Text-Based Adventure Game:
User will be prompted to navigate from the entrance to the area,
pick an opponent and select the right Weapon to kill the opponent
to win/lose the game.

Developers:
Abdulquayyum Yussuf
Markus Neil
Jadah McCullum
Ani Kimble
Jaren Lewis
"""
import random

# Rooms and locations
rooms = {
    'entrance': {
        'description': "\nCombat Arena\nLoading ......\nWelcome Warrior! You are at the entrance of the battleground. Ahead is the arena.",
        'items': [],
        'directions': ['north']
    },
    'arena': {
        'description': "You have entered the arena. Choose your opponent: Gladiator, Soldier, or Alien.",
        'items': [],
        'directions': ['north', 'south']
    },
    'gladiator_room': {
        'description': "You are facing the Gladiator. Choose your weapon: Grenade, Shield, or Laser.",
        'items': ['Grenade', 'Shield', 'Laser'],
        'directions': ['west', 'south']
    },
    'army_soldier_room': {
        'description': "You are facing an Soldier. Choose your weapon: Rifle, Shield, or Sword.",
        'items': ['Rifle', 'Shield', 'Sword'],
        'directions': ['west', 'south']
    },
    'alien_room': {
        'description': "You are facing an Alien. Choose your weapon: Laser, Blaster, or Sword.",
        'items': ['Laser', 'Blaster', 'Sword'],
        'directions': ['west', 'south']
    }
}

# Player's inventory starts empty
inventory = []

# Define Functions ----------------------------------------------
def describe_room(room):
    print(f"\n{rooms[room]['description']}")
    if rooms[room]['items']:
        print(f"Items available: {', '.join(rooms[room]['items'])}")
    print(f"Possible directions: {', '.join(rooms[room]['directions'])}")

def collect_item(room):
    """Collect an item from the current room, if any."""
    if rooms[room]['items']:
        item = input(f"\nChoose an item to collect from the room: {rooms[room]['items']}: ").capitalize()
        if item in rooms[room]['items']:
            inventory.append(item)
            rooms[room]['items'].remove(item)
            print(f"\n{item} collected! It's now in your inventory.")
        else:
            print(f"\n{item} is not available here. Try again.")
    else:
        print("No more items to collect here.")

def check_inventory():
    """Check player's inventory."""
    if inventory:
        print(f"\nYour inventory: {', '.join(inventory)}")
    else:
        print("\nYour inventory is empty.")

def battle(opponent):
    """Handle battle scenarios based on the opponent and the weapon choices."""
    print(f"\nYou are fighting a {opponent}.")

    if opponent == 'Gladiator':
        if 'Grenade' in inventory:
            print("You threw the grenade at the Gladiator, but he blocked it. Defeat!")
        elif 'Shield' in inventory:
            print("You blocked the Gladiator's attacks with the shield. Victory!")
        elif 'Laser' in inventory:
            print("You disintegrated the Gladiator with the laser. Victory!")

    elif opponent == 'Soldier':
        if 'Rifle' in inventory:
            print("You shot the Soldier. Victory!")
        elif 'Shield' in inventory:
            print("The soldier shot at you but your shield protected you. Victory!")
        elif 'Sword' in inventory:
            print("You charged with a sword but got shot. Defeat!")

    elif opponent == 'Alien':
        if 'Laser' in inventory:
            print("You overwhelmed the Alien's defenses with your laser. Victory!")
        elif 'Blaster' in inventory:
            print("You fired the blaster but the Alien absorbed it. Defeat!")
        elif 'Sword' in inventory:
            print("You slashed the Alien with the sword. Victory!")

def move_to_room(current_room, direction):
    """Handle movement between rooms."""
    if direction not in rooms[current_room]['directions']:
        print("You can't move in that direction.")
        return current_room

    if direction == 'north':
        if current_room == 'entrance':
            return 'arena'
        elif current_room == 'arena':
            opponent = input("\nChoose your opponent (Gladiator, Soldier, Alien): ").capitalize()
            if opponent == 'Gladiator':
                return 'gladiator_room'
            elif opponent == 'Soldier':
                return 'army_soldier_room'
            elif opponent == 'Alien':
                return 'alien_room'
            else:
                print("Invalid opponent choice!")
                return current_room
    elif direction == 'south':
        return 'arena' if current_room != 'entrance' else 'entrance'
    elif direction == 'west':
        return 'arena'

    return current_room

jester_encountered = False

def jester_riddle():
    """The Jester presents a riddle. Answer correctly to win a special item."""
    global jester_encountered
    if jester_encountered:
        return


    if random.randint(1, 5) == 1:
        print("\nA Jester appears and offers you a riddle!")
        print("Jester: 'Answer my riddle correctly, and I'll give you a special reward to aid you in your battles.'")
        print("Jester: 'What has keys but can't open locks?'")

        answer = input("\nYour answer: ").lower()

        if answer == "piano":
            print("\nJester: 'Correct! Take this potion of confidence, with this you can go into battle with no fear!(I know you probably wanted a sword or something but this is all i can afford)'")
            special_item = 'Potion of Confidence'
            inventory.append(special_item)
            print(f"\n{special_item} added to your inventory.")
        else:
            print("\nJester: 'Wrong answer! Better luck next time.' The Jester vanishes into thin air.")


        jester_encountered = True
#this segment above was done by Jaren Lewis. this is the bonus requirement #7


# ----------Main---------------
# Start the game
def game_loop():
    """Main game loop."""
    current_room = 'entrance'
    score = 0
    moves = 0

    while True:
        # Random chance for the Jester to appear in any room
        jester_riddle()

        describe_room(current_room)

        action = input("\nWhat's your next move? (move [direction], collect item, check inventory, exit): ").lower()

        if action.startswith('move'):
            direction = action.split()[1]
            new_room = move_to_room(current_room, direction)
            if new_room != current_room:
                current_room = new_room
                moves += 1

        elif action == 'collect item':
            collect_item(current_room)

        elif action == 'check inventory':
            check_inventory()

        elif action == 'exit':
            print(f"\nThanks for playing! You made {moves} moves and collected {len(inventory)} items.")
            break

        if current_room in ['gladiator_room', 'army_soldier_room', 'alien_room']:
            battle(current_room.split('_')[0].capitalize())
            score += 10

# --- Execute ------
game_loop()
```
